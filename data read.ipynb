```python
import json
import pandas as pd
```

```python
def json_get_length(path):
    with open(path, encoding="utf-8") as f:
        lines = f.readlines()
        return len(lines)
```

```python
def json_to_df(path, l):
    with open(path, encoding="utf-8") as f:
        dic = {}
        for i in range(l):
            line = f.readline()
            dic_new = dict(json.loads(line))
            if i == 0:
                dic = dic_new
                for key in dic:
                    dic[key] = [dic[key]]
            else:
                for key in dic:
                    dic[key].append(dic_new[key])
        f.close()
        dic = pd.DataFrame(dic)
        return dic
```
