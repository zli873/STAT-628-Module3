```python
import pandas as pd
from wordcloud import WordCloud
america_new_tip = pd.read_csv('american_new_tip.csv')  
```

```python
cloud = WordCloud(width=1440, height= 1080,max_words= 1000).generate(' '.join(america_new_tip['text'].astype(str)))
plt.figure(figsize=(20, 15))
plt.imshow(cloud)
plt.axis('off');
```

```python
from matplotlib import pyplot as plt
from wordcloud import WordCloud
from PIL import Image
import numpy as np
```

```python
img = Image.open('yelp.png')
img
img_array = np.array(img)
```

```python
yelp= WordCloud(
    background_color='white',
    width=1200,
    height=400,
    mask=img_array)
```

```python
yelp.generate(' '.join(america_new_tip['text'].astype(str)))
plt.figure(figsize=(20, 15))
plt.imshow(yelp)
plt.axis('off');
```
