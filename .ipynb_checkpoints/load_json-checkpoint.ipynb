```python
import json
import pandas as pd
```

```python
def json_to_df(path, lines = 100):
    with open(path, encoding="utf-8") as f:
        dic = {}
        for i in range(1,lines):
            line = f.readline()
            dic_new = dict(json.loads(line))
            if i == 1:
                dic = dic_new
                for key in dic:
                    dic[key] = [dic[key]]
            else:
                for key in dic:
                    dic[key].append(dic_new[key])
        f.close()
        dic = pd.DataFrame(dic)
        return dic
```

```python
review = json_to_df("review.json")
review
```

```python
tip = json_to_df("tip.json")
tip
```
